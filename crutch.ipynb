```python
inst['school_type'] = 2
```

```python
lm = lambda ll: [el for lst in ll for el in lst]
```

```python
allgroups = lm(inst.groups) + lm(ptu.groups)
```

```python
allgroups = list(set(allgroups))
```

```python
y = 0
groupsdict = {}
for x in allgroups:
    groupsdict[x] = y
    y += 1
```

```python
gti = lambda x: [groupsdict[el] for el in x]
```

```python
def gtb(y):
    gti = lambda x: [groupsdict[el] for el in x]
    otv = [False] * len(groupsdict)
    y = gti(y)
    for x in y:
        otv[x] = True
    otv = np.array(otv, dtype = np.bool_)
    return otv
```

```python
tmp = inst.groups.copy()
```

```python
inst.groups = inst.groups.map(gtb)
```
