```python
result = requests.get(api + 'database.getSchools?q=колледж&city_id=2&count=1000&v=5.69' + app_token)
json_result = result.json()
```

```python
data = {}
```

```python
kk = json_result['response']['items']
```

```python
for item in tqdm.tqdm_notebook(kk):
    time.sleep(1)
    sers = requests.get(api + 'users.search' + '?count=1000&school=' + str(item['id']) + token).json()
    data[item['id']] = sers['response'][1:]
    #print(sers['response'][1:])
```

## Засовываем в пандас

```python
data2 = {'first_name': [], 'last_name': [], 'uid': [], 'school': [], 'school_type': []}

for key, val in data.items():
    for info in val:
        data2['first_name'].append(info['first_name'])
        data2['last_name'].append(info['last_name'])
        data2['uid'].append(info['uid'])
        data2['school'].append(key)
        data2['school_type'].append(1)
```
