```python
# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

# Load pandas
import pandas as pd

# Load numpy
import numpy as np
import pickle
```

```python
with open('spare_HS_SPB.pickle', 'rb') as f:
    inst = pickle.load(f)
```

```python
with open('spare_PTU_SPB.pickle', 'rb') as f:
    ptu = pickle.load(f)
```

```python
inst
```

```python
tmp1 = inst.copy()
tmp2 = ptu.copy()
```

```python
tmp1 = tmp1.drop(['religion','uid','first_name','last_name'], axis = 1)
tmp2 = tmp2.drop(['religion','uid','first_name','last_name'], axis = 1)
```

```python
tmp3 = pd.concat([tmp1, tmp2]).sample(frac = 1)
```

```python
np.asmatrix(tmp3.groups).shape
```

```python
clf = RandomForestClassifier(max_depth = 3, n_jobs = -1)
```

```python
tmp3.head()
```

```python
from tqdm import tqdm_notebook
```

```python
import random
```

```python
tmp, tmp2 = [], []
tr, te = 9801, 4201
for x in tqdm_notebook(tmp3.head(tr).iterrows()):
    x = x[1]
   
    tmp += list(x['groups']) + [x['alchol']] + [x['has_photo']] + [x['life_main']] + [x['people_main']] + [x['political']] + [x['relation']] + [x['smoking']]
    
    tmp += [x['has_photo']] + [x['sex']]
    
    tmp2.append(x['school_type'])

tmp = np.array(tmp).reshape(-1, 186057)
tmp2 = np.array(tmp2)

clf.fit(tmp, tmp2)
```

```python
tmp1, tmp4 = [], []

for x in tqdm_notebook(tmp3.tail(te).iterrows()):
    x = x[1]
   
    tmp1 += list(x['groups']) + [x['alchol']] + [x['has_photo']] + [x['life_main']] + [x['people_main']] + [x['political']] + [x['relation']] + [x['smoking']]
    
    tmp1 += [x['has_photo']] + [x['sex']]
    
    tmp4.append(x['school_type'])

tmp1 = np.array(tmp).reshape(-1, 186057)
tmp4 = np.array(tmp2)

metrics.accuracy_score(tmp4, clf.predict(tmp1))
```
